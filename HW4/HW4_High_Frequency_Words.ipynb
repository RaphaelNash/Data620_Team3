```python
import nltk
import nltk.corpus.reader as r
from nltk import word_tokenize
from nltk import FreqDist
import string

%matplotlib inline
```

**Choose a corpus of interest.**

I choose negative movie reviews in the (polarity v2 dataset) from the following website:  http://www.cs.cornell.edu/people/pabo/movie-review-data/


```python
negative_reviews = r.PlaintextCorpusReader('review_polarity/txt_sentoken/neg/',  r'.*\.txt')
raw = negative_reviews.raw()
raw =  " ".join("".join([" " if ch in string.punctuation else ch for ch in raw]).split())
raw = raw.lower()
```

```python
words = nltk.word_tokenize(raw)
```

```python
words
```

**How many total unique words are in the corpus?**

```python
total_words = len(words)
total_words
```

**Taking the most common words, how many unique words represent half of the total words in the corpus?**

```python
freq_dist = FreqDist(words)
len(freq_dist)

unique_words_lt_50 = 0 
total_words_seen = 0 
for v in sorted(freq_dist.values(), reverse=True):
    unique_words_lt_50 = unique_words_lt_50+ 1
    total_words_seen = total_words_seen + v
 
    if total_words_seen > total_words * .5:
        break
unique_words_lt_50

```

**Identify the 200 highest frequency words in this corpus.**

```python
freq_dist.most_common(200)
```

```python
freq_dist.plot(200,cumulative=False)
```

**Does the observed relative frequency of these words follow Zipf’s law? Explain.***

Yes this does seem to follow Zipf's law pretty closely.  As you can see from the graph above each word seems to have a 50% less frequency than the previous word
